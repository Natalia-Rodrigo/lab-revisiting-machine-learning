```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
display(data)
```

```python
data.isna().sum()
```

```python
drop_list = []
drop_list.extend(['OSOURCE','ZIP'])
print(drop_list)
```

```python
def list_col_null_percentage(df: pd.DataFrame, threshold: float=0.25):
    
    null_df_def = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
    null_df_def.columns = ['col_name', 'null_percentage']

    col_above_thr = null_df_def[null_df_def['null_percentage']>threshold]
    drop_col_list = list(col_above_thr['col_name'])

    return drop_col_list
```

```python
col_85_nan = list_col_null_percentage(df=data, threshold=0.85)
print(col_85_nan)
```

```python
drop_list.extend(col_85_nan)
print(drop_list)
```

```python
data.drop(columns=drop_list, inplace=True)
```
